```python
from __future__ import division
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import scipy.io.wavfile as scw
import scipy.signal as scs
from IPython.display import Audio
import numpy as np
```

```python
fs, wave = scw.read('fluteircam.wav')
print('Data:', wave)
print('Sampling rate:', fs)
print('Audio length:', wave.size/fs, 'seconds')
print('Lowest amplitude:', min(wave))
print('Highest amplitude:', max(wave))

Audio('fluteircam.wav')
```

```python
def plotwave(fs, signal, maxf=None, depF=None):
    """Visualize (a segment of) a wave file."""
    # maxf = maximum number of frames
    frames = scipy.arange(signal.size)   # x-axis
    if maxf:
        plt.figure(figsize=(16, 6))
        plt.plot(frames[depF:maxf], signal[depF:maxf])
        plt.xticks(scipy.arange(depF, maxf, 0.5*fs), scipy.arange(depF, maxf/fs, 0.5))
        plt.show()
    else:
        plt.figure(figsize=(16, 6))
        plt.plot(frames, signal)
        plt.xticks(scipy.arange(0, signal.size, 0.5*fs), scipy.arange(0, signal.size/fs, 0.5))
        plt.show()
```

```python
plotwave(fs, wave, 30000, 29000)
```

```python
N=len(wave)
xf = np.linspace(-1.0/2.0*fs, 1.0/2.0*fs, N)
yf = scipy.fftpack.fft(wave)
yf = np.fft.fftshift(yf)
fig, ax = plt.subplots()

ax.plot(xf, 1.0/N *np.abs(yf) )
plt.xlim(0,5000)
plt.title('FFT du signal')

plt.grid()
```

```python
s=wave[int(0.3*N/14):int(13.5*N/14)]
 
# Plot the signal read from wav file
plt.figure(figsize=(16, 3))
plt.title('Spectrogram of a wav file with flute music')
plt.plot(s)
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.figure(figsize=(16, 7)) 
plt.specgram(s,Fs=fs) 
plt.xlabel('Time')
plt.ylabel('Frequency')
 
plt.show()
```
